```python
import pycaret.classification as pycc
import pickle
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
list_col_exc =["PassengerId", "Ticket"] + ["Name"]
```

```python
list_col_exc
```

```python
pycc.setup(df, target = "Survived", train_size=0.75, silent = True, use_gpu = True, ignore_features=(list_col_exc))
```

```python
best_clf = pycc.compare_models()
tune_model = pycc.tune_model(best_clf, choose_better = True)
pycc.finalize_model(tune_model)
```

```python
tune_model.predict_model("test.csv")
```
